### 1) Check Earlier and Latest date of data

```python
query = 'select \
        min(created_at) as Earlier_date , \
        max(created_at) as lastest_date from website_sessions'
my_cursor.execute(query)
pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
```

### 2) Check From Where the bulk of website sessions are coming and breakdown by utm-source , utm-campaign and refering domain.Take data till 11 April 2012

```python
query= "select utm_source ,\
             utm_campaign , \
             http_referer , \
             count(distinct website_session_id) as count  \
        from website_sessions \
        where created_at<= '2012-04-11' \
        group by 1,2,3 order by count desc "
my_cursor.execute(query)
pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
```

### 3) Get Traffic Source Conversion Rate. Take data till 13 April 2012 of gsearch nonbrand campaign.

```python
query =  "select count(website_sessions.website_session_id) as sessions , \
		    count(orders.order_id) as orders,\
            count(orders.order_id) / count(website_sessions.website_session_id) as Conv_Rate\
	from website_sessions left join orders on website_sessions.website_session_id = orders.website_session_id \
        where website_sessions.created_at <='2012-04-13' and website_sessions.utm_source = 'gsearch' and website_sessions.utm_campaign = 'nonbrand'"
my_cursor.execute(query)
pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
```

### 4) Show Gsearch Nonbrand trended session volume by week till 09 May 2012

```python
query = "select \
            date(created_at) as Week_start_date, \
		    count(website_session_id) as sessions \
		from website_sessions  \
        where website_sessions.created_at <='2012-05-09' and website_sessions.utm_source = 'gsearch' and website_sessions.utm_campaign = 'nonbrand'\
        group by week(created_at)"
my_cursor.execute(query)
x =pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
x
```

### 5) Get Conversion rate from session to orders by device_type till 10 may 2012

```python
query = "select device_type , \
		count(website_sessions.website_session_id) as sessions, \
        count(orders.order_id) as orders ,\
        count(orders.order_id)  / count(website_sessions.website_session_id) as conversion_rate\
    from website_sessions left join orders on website_sessions.website_session_id = orders.website_session_id \
    where website_sessions.created_at <='2012-05-10' and website_sessions.utm_source = 'gsearch' and website_sessions.utm_campaign = 'nonbrand' \
    group by device_type"
my_cursor.execute(query)
pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
```

### 6) Get most viewed website pages ranked by session volume till 08 June 2012

```python
query = "select \
                website_pageviews.pageview_url , \
                count(website_sessions.website_session_id) as sessions \
        from website_sessions left join website_pageviews on website_sessions.website_session_id = website_pageviews.website_session_id \
        where website_sessions.created_at <='2012-06-08' \
        group by website_pageviews.pageview_url \
        order by sessions desc"
my_cursor.execute(query)
pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
```

## 7) Pull Overall Sessions and Order Volume trended by quarter From April 01 , 2012 to December 31,2014

```python
query = "select year(website_sessions.created_at) as years,\
                quarter(website_sessions.created_at) as Quarters,\
                count(website_sessions.website_session_id) as Sessions, \
                count(orders.order_id) as orders \
                from website_sessions left join orders on website_sessions.website_session_id = orders.website_session_id \
                where website_sessions.created_at between '2012-04-01' and '2014-12-31' \
                group by 1,2 order by 1,2  "
my_cursor.execute(query)
pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
```

## 8) Get quarterly figured since website launch , Session to order Conversion rate , revenue per order , reveue per session

```python
query = " select 	year(website_sessions.created_at) as years, \
		quarter(website_sessions.created_at) as Quarters, \
		count(website_sessions.website_session_id) as Sessions, \
        count(orders.order_id) as orders , \
        count(orders.order_id) / count(website_sessions.website_session_id) as Coversion_Rate , \
        sum(orders.price_usd)/count(orders.order_id)  as revenue_per_order,\
        sum(orders.price_usd)/count(website_sessions.website_session_id)  as revenue_per_session\
        from website_sessions left join orders on website_sessions.website_session_id = orders.website_session_id \
        group by 1,2 order by 1,2 "
my_cursor.execute(query)
pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
```

## 9) Pull Quarterly View of orders from Gsearch Nonbrand , Bsearch nonBrand , Brand Search Overall, organic Search and Direct type in 

```python
query = "select year(website_sessions.created_at) as years , \
		quarter(website_sessions.created_at) as quarters, \
        count(distinct case when utm_source = 'gsearch'  and utm_campaign = 'nonbrand' then orders.order_id else null end) as Gsearch_NonBrand_Orders, \
        count(distinct case when utm_source ='bsearch' and utm_campaign = 'nonbrand' then orders.order_id else null end) as Bsearch_NonBrand_Orders , \
        count(distinct case when utm_campaign = 'brand' then orders.order_id else null end) as Overall_Brand_Orders , \
        count(distinct case when utm_source is null and http_referer is not null then orders.order_id else null end) as Organic_Search_Orders , \
        count(distinct case when utm_source is null and http_referer is null then orders.order_id else null end) as Direct_Type_in_Orders  \
		from website_sessions left join orders on website_sessions.website_session_id = orders.website_session_id \
        group by 1,2 "
my_cursor.execute(query)
pd.DataFrame(my_cursor.fetchall(), columns = my_cursor.column_names)
```
